```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load data
df = pd.read_csv('/content/drive/MyDrive/Data Mining/credit_data_sample.csv',
sep=';')

features = ['Age', 'Credit_amount', 'Duration',
'Employment_since']
X = df[features]

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method
inertia = []
K = range(1, 10)
for k in K:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X_scaled)
  inertia.append(kmeans.inertia_)

# Plot hasil elbow
plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.title('Metode Elbow untuk Menentukan k')
plt.grid(True)
plt.show()
```

```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load data
df = pd.read_csv('/content/drive/MyDrive/Data Mining/credit_data_sample.csv',
sep=';')

features = ['Age', 'Credit_amount', 'Duration',
'Employment_since']
X = df[features]

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Buat model KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
# Lihat hasil clustering
print(df.head())

# Visualisasi hasil clustering
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Credit_amount'],
y=df['Duration'], hue=df['Cluster'],palette='viridis')
plt.title('Cluster Calon Kreditur Berdasarkan Jumlah Kredit dan Durasi')
plt.xlabel('Credit Amount')
plt.ylabel('Duration')
plt.grid(True)
plt.show()
```

```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load data
df = pd.read_csv('/content/drive/MyDrive/Data Mining/credit_data_sample.csv',
sep=';')

features = ['Age', 'Credit_amount', 'Duration',
'Employment_since']
X = df[features]

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 7a. Hitung Silhouette Score
silhouette_avg = silhouette_score(X_scaled,df['Cluster'])
print(f"\nSilhouette Score: {silhouette_avg:.3f}")
# 7b. Tampilkan inertia akhir
print(f"Inertia (k=3): {kmeans.inertia_:.3f}")
# Step 8  Visualisasi hasil clustering (2 fitur:Credit_amount vs Duration)
plt.figure(figsize=(8,6))
sns.scatterplot(
  x=df['Credit_amount'],
  y=df['Duration'],
  hue=df['Cluster'],
  palette='viridis',
  s=100
)
plt.title('Cluster Calon Kreditur Berdasarkan Credit Amount dan Duration')
plt.xlabel('Credit Amount')
plt.ylabel('Duration')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Load data
df = pd.read_csv('/content/drive/MyDrive/Data Mining/Iris.csv',
sep=',')

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
'PetalWidthCm']
X = df[features]

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Menggunakan DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Visualisasi hasil clustering
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels,
cmap='viridis')
plt.title("Hasil DBSCAN Clustering")
plt.show()
```
