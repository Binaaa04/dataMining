```python
import pandas as pd

#data dalam bentuk dictionary
data ={
    'land area/luas tanah (m2)' : [100, 150 ,80 ,120,200,90,130,110,140,95],
    'total rooms/ jumlah kamar' : [3,4,2,3,5,2,3,3,4,2],
    'distance to center city/ jarak ke pusat kota(km)': [5,10,2,8,15,7,12,6,9,4],
    'prices/ harga (juta rupiah)': [500,450,400,350,700,300,380,480,420,320],
}
#membuat data frames
df = pd.DataFrame(data)
print(df)

#menghitung matriks korelasi
correlation_matrix = df.corr()
#menampilkan korelasi antara fitur dan target
print(correlation_matrix['prices/ harga (juta rupiah)'])

#add power interruptions data
df ['power interruptions/ gangguan listrik (kwh)'] = [5, 4, 5, 3, 2, 5, 3, 4, 2, 5]
print("dataframe after adding power interruptions")
print(df)

#menghitung matriks korelasi
correlation_matrix = df.corr()
# Menampilkan korelasi antara kolom-kolom dengan kolom 'prices/ harga (juta rupiah)'
print("\nKorelasi antara kolom-kolom dengan 'prices/ harga (juta rupiah)':")
print(correlation_matrix['prices/ harga (juta rupiah)'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#visualisasi matriks korelasi menggunakan heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Fitur')
plt.show()
#
```

```python
import pandas as pd

# parameter slovin
N = 100  # ukuran populasi (100)
e = 0.05  # tingkat kesalahan 5%

# hitung ukuran sampel menggunakan rumus Slovin
n = N / (1 + N * e**2)
print(f"Ukuran sampel yang diperlukan: {n:.0f}")  # Membulatkan ke angka bulat

# buat dataframe dummy dengan 100 data
data = {'value': range(1, 101)}
df = pd.DataFrame(data)

# ambil sampel acak dari dataframe
sample_df = df.sample(n=int(n), random_state=42)  # random state untuk hasil yang konsisten

# tampilkan sample
print("Sample yang diambil:")
print(sample_df)
```

```python
import pandas as pd
import numpy as np

# Parameter Slovin
N = 1000  # Ukuran populasi
e = 0.05  # Tingkat kesalahan 5%

# Hitung ukuran sampel
n = N / (1 + N * e**2)
print(f"Ukuran sampel yang diperlukan: {n:.0f}")  # Membulatkan ke angka bulat

# Buat DataFrame dummy dengan 1000 data (5 kolom)
np.random.seed(42)  # Agar hasil acak konsisten
data = {
    'ID': range(1, 1001),  # ID dari 1 sampai 1000
    'Nama': [f'User{i}' for i in range(1, 1001)],  # Nama User1, User2, ...
    'Umur': np.random.randint(18, 60, 1000),  # Umur antara 18-60 tahun
    'Pendapatan': np.random.randint(3000000, 10000000, 1000),  # Gaji random
    'Status': np.random.choice(['Single', 'Menikah'], 1000)  # Status acak
}
df = pd.DataFrame(data)

# Ambil sampel acak dari DataFrame
sample_df = df.sample(n=int(n), random_state=42)

# Tampilkan sample yang diambil
print("Sample yang diambil:")
print(sample_df)
```

```python
import pandas as pd
import seaborn as sns

# Load dataset Titanic dari seaborn
df = sns.load_dataset('titanic')

# Bersihkan data (hapus baris yang ada nilai kosong)
df_cleaned = df.dropna()

# Parameter Slovin
N = len(df_cleaned)  # Total populasi
e = 0.05  # Tingkat kesalahan 5%

# Hitung jumlah sampel
n = N / (1 + N * e**2)
n = int(n)  # Dibulatkan ke bawah

# Ambil sampel secara acak
sample_df = df_cleaned.sample(n=n, random_state=42)

# Tampilkan ukuran populasi & sampel
print(f"Ukuran Populasi: {N}")
print(f"Ukuran Sampel (Slovin): {n}")
print(sample_df.head())

```
