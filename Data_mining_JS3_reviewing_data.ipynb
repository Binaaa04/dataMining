```python
import seaborn as sns  # Import Seaborn untuk visualisasi data
import matplotlib.pyplot as plt  # Import Matplotlib untuk mengatur ukuran dan tampilan grafik

df = sns.load_dataset('diamonds')  # Load dataset 'diamonds' ke dalam DataFrame

plt.figure(figsize=(8, 6))  # Mengatur ukuran grafik menjadi 8x6 inci

sns.countplot(x='cut', data=df, palette='Set1')
# Membuat grafik batang untuk melihat jumlah berlian berdasarkan kualitas potongan ('cut')
# Menggunakan warna dari palet 'Set1' agar lebih menarik

plt.title('Distribution Diamond Cut', fontsize=16)  # Menambahkan judul dengan ukuran font 16
plt.xlabel('Diamond Cut', fontsize=14)  # Menambahkan label sumbu X dengan ukuran font 14
plt.ylabel('Total diamond', fontsize=14)  # Menambahkan label sumbu Y dengan ukuran font 14

plt.show()  # Menampilkan grafik
```

```python
import seaborn as sns  # Import Seaborn untuk mengambil dataset

df = sns.load_dataset('diamonds')  # Load dataset 'diamonds' ke dalam DataFrame

cut_counts = df['cut'].value_counts()
# Menghitung jumlah berlian berdasarkan jenis potongan ('cut')
# Hasilnya adalah jumlah masing-masing kategori seperti 'Fair', 'Good', dll.

print("Total diamond in each cut:")  # Menampilkan teks sebagai keterangan
print(cut_counts)  # Menampilkan jumlah berlian untuk setiap jenis potongan
```
